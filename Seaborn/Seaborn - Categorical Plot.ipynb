```python
import seaborn as sns
import matplotlib.pyplot as plt
%matplotlib inline
import pandas as pd
sns.set(style="ticks", color_codes=True)
```

```python
tips = pd.read_csv('./data/tips.csv')
```

```python
tips.head(), tips.shape
```

## Categorical scatterplots

*__auto adds small amount of random â€œjitter"__*

```python
sns.catplot(x="day", y="total_bill", jitter=False, data=tips);
```

**_without jitter_**

```python
sns.catplot(x='day', y='total_bill', data=tips)
```

**_prevents points from overlapping (only works well for relatively small datasets)_**

```python
sns.catplot(x="day", y="total_bill", data=tips, kind='swarm');
```

**_3 variables_**

```python
sns.catplot(x="day", y="total_bill", data=tips, kind='swarm', hue='sex');
```

**_excluding a value_**

```python
sns.catplot(x="size", y="total_bill", data=tips.query('size != 3'), kind='swarm');
```

**_controlling the order_**

```python
sns.catplot(x='smoker', y='tip', data=tips, order=['Yes','No'])
```

**_on vertical axis (just swap the x and y)_**

```python
sns.catplot(x='total_bill', y='day', data=tips, kind='swarm', hue='time')
```

## Distributions within categories

```python
sns.catplot(x='day', y='total_bill', data=tips, kind='box')
```

```python
sns.catplot(x='day', y='tip', data=tips, kind='box', hue='smoker')
```

**_for only specific values from a column_**

```python
tips['weekend'] = tips['day'].isin(['Sat', 'Sun'])
sns.catplot(x='day', y='tip', data=tips, kind='box', hue='weekend')
```

```python
diamonds = pd.read_csv('./data/seaborn-diamonds.csv')
diamonds.head(), diamonds.shape
```

**_best suited for larger datasets_**

```python
sns.catplot(x='color', y='price', data=diamonds.sort_values('color'), kind='boxen')
```

```python
titanic = pd.read_csv('./data/seaborn-titanic.csv')
titanic.head(), titanic.shape
```

```python
sns.catplot(x='sex', y='survived', data=titanic, kind='bar', hue='class')
```

**_1 variable with count (similar to value_counts)_**

```python
sns.catplot(x='deck', kind='count', data=titanic)
```

```python
sns.catplot(x='deck', kind='count', data=titanic, hue='class')
```

```python
sns.catplot(x='sex', y='survived', kind='point', data=titanic, hue='class')
```

**_control the size and shape_**

```python
plt.figure(figsize=(15, 12))
ax = sns.barplot(y='survived',x='alone',data=titanic);
```

**_multiple relationships with facets_**

```python
sns.catplot(y='tip', x='day', data=tips, hue='smoker',
           kind='swarm', col='time', aspect=.6)
```

```python
sns.catplot(y='survived', x='fare', data=titanic.query('fare > 0'),
           kind='box', row='class', aspect=4, orient='h', height=1.5)
```
