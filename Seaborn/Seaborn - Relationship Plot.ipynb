```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
sns.set(style="darkgrid")
```

```python
tips = pd.read_csv('./data/seaborn-tips.csv')
tips.head(), tips.shape
```

## scatter plots

**_In relplot ->  kind='scatter' is the default kind_**

**_2 variables (numeric, numerid)_**

```python
sns.relplot(x='total_bill', y='tip', data=tips)
```

**_3 variables (numeric, numeric, categorical)_**

```python
sns.relplot(x='total_bill', y='tip', data=tips, hue='smoker')
```

```python
sns.relplot(x='total_bill', y='tip', data=tips, hue='sex', style='sex')
```

**_4 variables (numeric, numeric, categorical, categorical)_**

```python
sns.relplot(x='total_bill', y='tip', data=tips, hue='smoker', style='sex')
```

**_(numeric, numeric, numeric) (color changes)_**

```python
sns.relplot(x="total_bill", y="tip", hue="size", data=tips)
```

**_change color_**

```python
sns.relplot(x="total_bill", y="tip", hue="size", data=tips, palette='ch:r=-.5,l=.75')
```

**_change size of points_**

```python
sns.relplot(x="total_bill", y="tip", size="size", data=tips)
```

**_change size of size_**

```python
sns.relplot(x="total_bill", y="tip", data=tips, size="size", sizes=(15,200))
```

```python
fmri = pd.read_csv('./data/seaborn-fmri.csv')
fmri.head(), fmri.shape
```

**_multiple measurements for the same value of the x variable_**

**_seaborn is to aggregate the multiple measurements at each x value by plotting the mean and the 95% confidence interval around the mean_**

```python
sns.relplot(x='timepoint', y='signal', data=fmri, kind='line')
```

**_to disable it_**

```python
sns.relplot(x='timepoint', y='signal', data=fmri, kind='line', ci=None)
```

**_spread of distribution_**

**_plotting the standard deviation instead of a confidence interval_**

```python
sns.relplot(x='timepoint', y='signal', data=fmri, kind='line', ci='sd')
```

**_3 variable, changes same as in scatter plot (hue, size, style)_**

```python
sns.relplot(x='timepoint', y='signal', data=fmri, kind='line', hue='event')
```

```python
sns.relplot(x='timepoint', y='signal', data=fmri, kind='line', hue='event', style='event')
```

**_4 variables_**

```python
sns.relplot(x='timepoint', y='signal', data=fmri, kind='line', hue='region', style='event', dashes=False, markers=True)
```

## multiple relationships with facets

```python
sns.relplot(x='total_bill', y='tip', data=tips, hue='smoker', col='time')
```

```python
sns.relplot(x='timepoint', y='signal', data=fmri, kind='line', estimator=None,
            hue='subject', 
            col='region', row='event')
```

**_examine effects across many levels of a variable_**

**_facet that variable on the columns and then wrap the facets into the rows_**

```python
sns.relplot(x='timepoint', y='signal', data=fmri.query('region == "frontal"'), kind='line',
            hue='event', style='event', 
            col='subject', col_wrap=5,
            aspect=.75, height=3)
```
